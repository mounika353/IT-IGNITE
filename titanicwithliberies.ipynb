```python

women = df[df['sex'] == 'female']
print(f'Female passengers: {len(women)}')


first_class_survivors = df[(df['pclass'] == 1) & (df['survived'] == 1)]
print(f'1st class survivors: {len(first_class_survivors)}')


high_class = df[df['pclass'].isin([1, 2])]
print(f'1st or 2nd class: {len(high_class)}')


young_survivors = df.query('age < 18 and survived == 1')
print(f'Surviving children: {len(young_survivors)}')
young_survivors[['sex','age','pclass','fare']].head()
```

```python
print('Top 5 by fare:')
print(df.sort_values('fare', ascending=False)[['sex','age','pclass','fare']].head())

print('\nSort by class asc then fare desc:')
print(df.sort_values(['pclass','fare'], ascending=[True,False])
        [['pclass','fare','sex','age']].head(8))
```

```python
print('Survival rate by Sex:')
print(df.groupby('sex')['survived'].mean().round(3))

print('\nSurvival rate by Class:')
print(df.groupby('pclass')['survived'].mean().round(3))

print('\nFare stats by Class:')
print(df.groupby('pclass')['fare'].agg(['mean','median','std','max']).round(2))

print('\nSurvival by Sex x Class:')
print(df.groupby(['pclass','sex'])['survived'].mean().round(3).unstack())
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
result = pd.DataFrame({'Count': missing, 'Pct': missing_pct})
print(result[result['Count'] > 0])
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
print(f'Age nulls after fill   : {df["age"].isnull().sum()}')

df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
print(f'Embark nulls after fill: {df["embark_town"].isnull().sum()}')

print(f'Final shape: {df.shape}')
```

```python
df['fare_per_person'] = df['fare'] / df['sibsp'].replace(0, 1)


bins   = [0, 12, 18, 35, 60, 100]
labels = ['Child','Teen','Young Adult','Adult','Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
print('Age group distribution:')
print(df['age_group'].value_counts().sort_index())


df['fare_category'] = df['fare'].apply(
    lambda x: 'Low' if x < 15 else ('Medium' if x < 50 else 'High')
)
print('\nFare category:')
print(df['fare_category'].value_counts())

df[['age','age_group','fare','fare_category']].head(8)
```

```python
print('PANDAS PRACTICE TASKS')
print('=' * 40)

print('T1 - Survival rate by embarkation town:')
print(df.groupby('embark_town')['survived'].mean().round(3))

print('\nT2 - Avg fare survived vs died:')
print(df.groupby('survived')['fare'].mean().round(2))

print('\nT3 - Top 3 embarkation towns:')
print(df['embark_town'].value_counts().head(3))

women_surv = df[df['sex']=='female']['survived'].mean()
print(f'\nT4 - Women survival rate: {women_surv:.1%}')

print('\nT5 - Mean fare by class x sex:')
print(pd.pivot_table(df, values='fare', index='pclass', columns='sex', aggfunc='mean').round(2))
```

```python
fare_cats = ['Low', 'Medium', 'High']
avg_ages  = [df[df['fare_category']==c]['age'].mean()  for c in fare_cats]
avg_fares = [df[df['fare_category']==c]['fare'].mean() for c in fare_cats]

fig, ax = plt.subplots(figsize=(9, 5))

ax.plot(fare_cats, avg_ages,  marker='o', linewidth=2.5, markersize=9,
        color='steelblue', label='Avg Age')
ax.plot(fare_cats, avg_fares, marker='s', linewidth=2.5, markersize=9,
        color='tomato', linestyle='--', label='Avg Fare (GBP)')

for i, age in enumerate(avg_ages):
    ax.annotate(f'{age:.1f}', (fare_cats[i], age),
                textcoords='offset points', xytext=(0,10), ha='center')

ax.set_title('Average Age & Fare by Fare Category', fontsize=15, fontweight='bold')
ax.set_xlabel('Fare Category')
ax.set_ylabel('Value')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.4, linestyle='--')
plt.tight_layout()
plt.show()
print('Insight: Higher-fare passengers tend to be older (1st class).')
```

```python
surv_class = df.groupby('pclass')['survived'].value_counts().unstack()

fig, ax = plt.subplots(figsize=(9, 5))
x, w = np.arange(3), 0.35

b1 = ax.bar(x - w/2, surv_class[0], w, label='Did Not Survive',
            color='tomato', edgecolor='white')
b2 = ax.bar(x + w/2, surv_class[1], w, label='Survived',
            color='mediumseagreen', edgecolor='white')

for bar in list(b1) + list(b2):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+2,
            str(int(bar.get_height())), ha='center', fontsize=10)

ax.set_title('Survival Count by Passenger Class', fontsize=15, fontweight='bold')
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Count')
ax.set_xticks(x)
ax.set_xticklabels(['1st','2nd','3rd'])
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.4, linestyle='--')
plt.tight_layout()
plt.show()
```
